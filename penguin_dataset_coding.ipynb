```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_iphones(base_url, max_pages):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    all_products = []

    for page in range(1, max_pages + 1):
        print(f"Scraping page {page}...")
        url = f"{base_url}&page={page}"  # Append the page number
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            product_containers = soup.find_all('div', class_='_1AtVbE')

            for product in product_containers:
                try:
                    name = product.find('div', class_='_4rR01T').text.strip()
                except AttributeError:
                    name = None

                try:
                    price = product.find('div', class_='_30jeq3').text.strip().replace("₹", "").replace(",", "")
                    price = int(price) if price else None  # Convert to integer
                except AttributeError:
                    price = None

                try:
                    rating = product.find('div', class_='_3LWZlK').text.strip()
                    rating = float(rating) if rating else None  # Convert to float
                except AttributeError:
                    rating = None

                try:
                    num_reviews = product.find('span', class_='_2_R_DZ').text.strip().split("&")[0]
                    num_reviews = num_reviews.replace(",", "").split()[0]
                    num_reviews = int(num_reviews) if num_reviews.isdigit() else None  # Convert to integer
                except AttributeError:
                    num_reviews = None

                availability = 'Available' if price else 'Out of Stock'

                if name and price:
                    all_products.append({
                        'Product Name': name,
                        'Price': price,
                        'Rating': rating,
                        'Number of Reviews': num_reviews,
                        'Availability': availability
                    })
        else:
            print(f"Failed to fetch page {page}")

        time.sleep(2)  # Pause to avoid overwhelming the server

    df = pd.DataFrame(all_products)
    return df
```

```python
flipkart_url = 'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy,4io'
df_iphones = scrape_iphones(flipkart_url, max_pages=5)

# Save the data to a CSV file
df_iphones.to_csv('iphones_data.csv', index=False)
print("Data successfully saved to iphones_data.csv")

# Display the first few rows
df_iphones.head()
```
